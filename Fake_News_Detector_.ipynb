```python
# stemming
ps = PorterStemmer()
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [ps.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter

# 1. Word Frequency Distribution for Real and Fake News:

# Assuming 'text' column contains the news content
def plot_word_frequency(df, label_value, label_name, top_n=20):
    """Plots the frequency distribution of the top N words for a specific label."""
    text = ' '.join(df[df['label'] == label_value]['text'].astype(str).tolist())
    words = text.lower().split()
    word_counts = Counter(words)
    top_words = word_counts.most_common(top_n)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=[word[0] for word in top_words], y=[word[1] for word in top_words])
    plt.title(f'Top {top_n} Most Frequent Words in {label_name} News')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Plot for real news (label 0)
plot_word_frequency(news_df, 0, 'Real')

# Plot for fake news (label 1)
plot_word_frequency(news_df, 1, 'Fake')

```
